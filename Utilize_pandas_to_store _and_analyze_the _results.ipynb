```python
#Step 3.1 Determine the outcome of each game, based on the final player totals.

#Addng a new column outcome
def outcome(Final_sum_P1,Final_sum_P2):
    
    if Final_sum_P1>Final_sum_P2:
        return "Player 1 wins"
    elif Final_sum_P1<Final_sum_P2:
        return "Player 2 wins"
    else:
        return "Tie game"
gmrslts['outcome'] = gmrslts.apply(lambda x: outcome(x['Final_sum_P1'],x['Final_sum_P2']),axis=1)
```

```python
#Step 3.2 
#Add new columns for the winning totals and missing values  for tie games. 
def outcome1(Final_sum_P1,Final_sum_P2):
    if Final_sum_P1>Final_sum_P2:
        return Final_sum_P1
    elif Final_sum_P1<Final_sum_P2:
        return Final_sum_P2
    else:
        return np.nan
gmrslts['Winning Total'] = gmrslts.apply(lambda x: outcome1(x['Final_sum_P1'],x['Final_sum_P2']),axis=1)
```

```python
def outcome2(Final_sum_P1,Final_sum_P2):
    if Final_sum_P1>Final_sum_P2:
        return Final_sum_P2
    elif Final_sum_P1<Final_sum_P2:
        return Final_sum_P1
    else:
        return np.nan
gmrslts['Losing Total'] = gmrslts.apply(lambda x: outcome2(x['Final_sum_P1'],x['Final_sum_P2']),axis=1)
```
